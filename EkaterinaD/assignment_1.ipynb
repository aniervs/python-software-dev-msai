```python
def is_prime(n):
    """
    Check if the number is prime or not.
    """
    if n <= 1:
        return False

    # Check divisibility from 2 up to the square root of n
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False  # n is divisible by i, so it's not prime

    return True  # If no divisors found, n is prime

v =[print(str(i) + " is prime: "+ str(is_prime(i))) for i in range (1,15)]
```

```python
def limit(input_generator, max_count):
    """
    Generator that returns not more than max_count values of the input_generator.
    """
    for _ in range(max_count):
        yield next(input_generator)

print(list(limit((x for x in range(10)), 8)))
```

## Exercise 5

Write a generator for an infinite sequence of numbers from the Pascal's triangle. The sequence look like this:
`1 1 1 1 2 1 1 3 3 1 1 4 6 4 1 1 5 10 10 5 1 1 6 15 20 15 6 1 1 7 21 35 35 21 7 1 1 8 28 56 70 56 28 8 1 1 9 36 84 126 126 84 36 9 1 ... '

Test it with a generator from the previous task)

```python
# your code here
def generate_pascals_triangle():
    row = [1]  # The first row of Pascal's Triangle is [1]
    while True:
        yield row
        # Generate the next row using the current row
        row = [1] + [row[i] + row[i+1] for i in range(len(row) - 1)] + [1]

for val in limit(generate_pascals_triangle(), 100):
  print(val, end=" ")
```

```python
import os

def files_sorted_by_size(path_to_dir):
    """
    Return a list of files in path_to_dir sorted by size.
    The same size files sorted alphabetically
    """
    # List all files in the given directory
    files = [f for f in os.listdir(path_to_dir) if os.path.isfile(os.path.join(path_to_dir, f))]

    # Sort the files by size and then alphabetically
    files.sort(key=lambda f: (os.path.getsize(os.path.join(path_to_dir, f)), f))

    return files
```

## Exercise 7

Write a `merge_sorter` generator that merges sorted sequences of integers.

The generator takes an arbitrary number of arguments. The argument can be any iterable, including another generator. It is guaranteed that each argument is a sequence of integers, sorted in non-decreasing order.

```python
def merge_sorter(*sequences):
    # Create an iterator for each sequence
    iterators = [iter(seq) for seq in sequences]
    # Initialize a list to store the first element from each iterator
    min_heap = []

    # Populate the heap with the first element from each iterator
    for idx, it in enumerate(iterators):
        try:
            first_elem = next(it)
            min_heap.append((first_elem, idx))
        except StopIteration:
            continue

    # While there are elements to process
    while min_heap:
        # Pop the smallest element (min-heap behavior is implicit with sorting)
        min_heap.sort()
        val, seq_idx = min_heap.pop(0)
        yield val

        # Try to get the next element from the same iterator
        try:
            next_elem = next(iterators[seq_idx])
            min_heap.append((next_elem, seq_idx))
        except StopIteration:
            continue

seq1 = [1, 5, 10]
seq2 = [1, 3, 2, 8, 7]

for n in merge_sorter(seq1, seq2):
    print(n)
```

## Exercise 8

Write the decorator `proler`, which, when calling a function, will store in its attributes (not to be confused with arguments) the time of its execution (in seconds, it can be fractional) and the number of recursive calls that occurred during execution. Name the attributes `last_time_taken` and `calls`.
It is forbidden to use global variables.
The decorator must behave in a decent manner, that is, it must not overwrite the function's documentation.

For tests write [Ackermann function](https://en.wikipedia.org/wiki/Ackermann_function)

```python
import time
import functools

def profiler(function):
    @functools.wraps(function)
    def wrapper(*args, **kwargs):
        if not hasattr(wrapper, 'calls'):
            wrapper.calls = 0

        wrapper.calls += 1

        start = datetime.now()
        result = function(*args, **kwargs)
        end = datetime.now()

        wrapper.last_time_taken = end - start

        return result
    return wrapper

# Ackermann function as an example of recursive function
@profiler
def ackermann(m, n):
    """Ackermann's function."""
    if m == 0:
        return n + 1
    elif m > 0 and n == 0:
        return ackermann(m - 1, 1)
    else:
        return ackermann(m - 1, ackermann(m, n - 1))

```

```python
result = ackermann(3, 6)
print("Result:", result)
print("Calls:", ackermann.calls)
print("Time taken:", ackermann.last_time_taken, "seconds")
```

## Exercise 9

Write the function `encode` that implements [run-length encoding](https://en.wikipedia.org/wiki/Run-length_encoding) algorithm

```python
def encode(sequence):
    if not sequence:
        return ""

    encoded = []
    count = 1

    for i in range(1, len(sequence)):
        if sequence[i] == sequence[i - 1]:
            count += 1
        else:
            encoded.append(sequence[i - 1] + str(count))
            count = 1

    # Append the last character and its count
    encoded.append(sequence[-1] + str(count))

    return ''.join(encoded)


encoded_str = encode("qqqqwwddwwdwdw")
print(encoded_str)
```

## Exercise 10

Write a decorator `visualizer` that takes a recursive function and will visualize the recursive calls that are made during the execution.
Consider using the `networkx` library or some other one in order to draw the recursion tree.
Test it with a recursive function that computes fibonacci.

```python
def fibonacci(n):
    if n == 0 or n == 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
import functools
import networkx as nx
import matplotlib.pyplot as plt

def visualizer(function):
    @functools.wraps(function)
    def wrapper(*args, **kwargs):
        wrapper.__calls += 1
        current = str(wrapper.__calls)
        wrapper.__graph.add_node(current, label=repr(args))  # Add node with the call arguments

        old_current = wrapper.__current
        wrapper.__current = current
        if old_current is not None:
            wrapper.__graph.add_edge(current, old_current)  # Add directed edge from previous call

        result = function(*args, **kwargs)

        wrapper.__current = old_current
        return result

    wrapper.__graph = nx.DiGraph()  # Directed graph for function call visualization
    wrapper.__calls = 0  # Number of calls
    wrapper.__current = None  # Current node (last call)

    # After the function finishes, plot the graph
    def plot_graph():
        pos = nx.spring_layout(wrapper.__graph)  # Layout for the graph
        labels = nx.get_node_attributes(wrapper.__graph, 'label')  # Get node labels
        nx.draw(wrapper.__graph, pos, with_labels=True, node_size=2000, node_color='green')
        nx.draw_networkx_labels(wrapper.__graph, pos, labels=labels)
        plt.show()

    wrapper.plot_graph = plot_graph  # Attach the plotting function to the wrapper
    return wrapper
```

```python
@visualizer
def fibonacci(n):
    if n == 0 or n == 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
value = fibonacci(5)
print("fibonacci(5) =", value)


fibonacci.plot_graph()
```

## Exercise 11

Now write a decorator `memoizer` that will do caching on the calls of a function (memoization).
Decorate fibonacci with the decorators `memoizer`, `profiler` and `visualizer` at the same time (not necessarily in that order). Test that they're working as one would expect.

```python
import functools

def memoizer(function):
    cache = {}  # Cache for storing results
    @functools.wraps(function)
    def wrapper(*args, **kwargs):
        if args in cache:  # If the result is cached, return it
            return cache[args]
        result = function(*args, **kwargs)  # Otherwise, compute the result
        cache[args] = result  # Store the result in the cache
        return result
    return wrapper
```

```python

@profiler
@visualizer
@memoizer
def fibonacci2(n):
    if n == 0 or n == 1:
        return n
    return fibonacci2(n - 1) + fibonacci2(n - 2)


value = fibonacci2(5)
print("fibonacci(5) =", value)

print("Result:", value)
print("Calls:", fibonacci2.calls)
print("Time taken:", fibonacci2.last_time_taken, "seconds")

fibonacci2.plot_graph()
```

```python
list_of_objects = [
    int,
    2,
    2.,
    None,
    object,
    str,
    str(2.),
    float('2.0'),
    'hello',
    dict,
    list,
    [dict],
    {1: []}
]

list_of_types = [
    int,
    float,
    object,
    str,
    dict,
    list
]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the list of objects and types
list_of_objects = [
    int,
    2,
    2.,
    None,
    object,
    str,
    str(2.),
    float('2.0'),
    'hello',
    dict,
    list,
    [dict],
    {1: []}
]

list_of_types = [
    int,
    float,
    object,
    str,
    dict,
    list
]

# Create a correspondence matrix (object x type)
correspondence_matrix = np.zeros((len(list_of_objects), len(list_of_types)))

# Fill the matrix using isinstance()
for i, obj in enumerate(list_of_objects):
    for j, t in enumerate(list_of_types):
        if isinstance(obj, t):
            correspondence_matrix[i, j] = 1

# Visualize the matrix using pcolormesh
plt.figure(figsize=(8, 6))
plt.pcolormesh(correspondence_matrix, cmap='coolwarm', edgecolors='k', linewidth=1)

# Set labels and title
plt.xticks(np.arange(0.5, len(list_of_types)), list(map(str, list_of_types)), rotation=45)
plt.yticks(np.arange(0.5, len(list_of_objects)), list(map(str, list_of_objects)))
plt.xlabel('Types')
plt.ylabel('Objects')
plt.title('Correspondence Matrix (Object  Type)')

# Show colorbar
plt.colorbar(label='1 = Is instance, 0 = Not instance')

plt.tight_layout()

plt.show()

correspondence_matrix
```
